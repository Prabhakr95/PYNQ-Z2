# CLAHE

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('C:\JN\chaubey/sp12/prabhakarsp12.jpg', cv2.IMREAD_GRAYSCALE)

# Check if the image is loaded successfully
if image is None:
    print("Error: Could not open or read the image.")
    exit()
```

```python
# Create a CLAHE object (Arguments: clipLimit, tileGridSize)
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
```

```python
# Apply CLAHE to the image
clahe_result = clahe.apply(image)
```

```python
# Display the original and equalized images
plt.figure(figsize=(10, 5))
```

```python
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
```

```python
plt.subplot(1, 2, 2)
plt.imshow(clahe_result, cmap='gray')
plt.title('Equalized Image')
```

```python
# Calculate histograms
hist_original = cv2.calcHist([image], [0], None, [256], [0, 256])
hist_equalized = cv2.calcHist([clahe_result], [0], None, [256], [0, 256])
```

```python
# Plot histograms
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(2, 2, 2)
plt.imshow(clahe_result, cmap='gray')
plt.title('Equalized Image')

plt.subplot(2, 2, 3)
plt.plot(hist_original, color='red')
plt.title('Original Image Histogram')
plt.xlabel('Pixel value')
plt.ylabel('Frequency')
plt.subplot(2, 2, 4)
plt.plot(hist_equalized, color='blue')
plt.title('Equalized Image Histogram')
plt.xlabel('Pixel value')
plt.ylabel('Frequency')
plt.show()
```

```python
from skimage.metrics import peak_signal_noise_ratio, structural_similarity


# Calculate PSNR
psnr_value = peak_signal_noise_ratio(image, clahe_result)

# Calculate SSIM
ssim_value, _ = structural_similarity(image, clahe_result, full=True)

# Display PSNR and SSIM values
print(f"PSNR: {psnr_value:.2f} dB")
print(f"SSIM: {ssim_value:.4f}")
```
